```python
import requests, bs4
import pandas as pd
```

```python
def scrape_links(page_url):
    r = requests.get(page_url)
    soup = bs4.BeautifulSoup(r.text, 'html.parser')
    link_holders = soup.select('h2 > a')
    if not link_holders:
        link_holders = soup.select('li > a')
    links = {link.text: link['href'] for link in link_holders}
    return links

def get_pages(start_page=1, stop_page=50):
    links = {}
    for page in range(start_page, stop_page+1):
        link = f'https://tooxclusive.com/main/naija-lyrics/page/{page}/'
#         if page==1:
#             link = f'https://tooxclusive.com/main/naija-lyrics/'
        l = scrape_links(link)
        if not l:
            break
        links.update(l)
    return links
```

```python
def get_lyric(link):
    r = requests.get(link)
    soup = bs4.BeautifulSoup(r.text, 'html.parser')
    selected = soup.select('div.hentry-content > p')[3:-1]
    if not selected:
        selected = soup.select('p')[2:-3]
    lyric = [x.text for x in selected]
    return '\n\n'.join(lyric)
```

```python
def get_df(start, stop):
    titles, lyrics = [], []
    links = get_pages(start, stop)
    for title, link in links.items():
        lyric = get_lyric(link)
        titles.append(title)
        lyrics.append(lyric)
        if len(lyric)<100:
            print(title, link)
    df = pd.DataFrame(index=None)
    df['Title'] = titles
    df['Lyric'] = lyrics
    return df
```

```python
df = get_df(1, 3)
# this will get all the lyrics from page 1 to page 3 and save it to a pandas dataframe
```
